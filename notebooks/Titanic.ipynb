```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("D://Practice//TitanicDatasetDataCleaning//train.csv")
display(data)
```

```python
data.count()
```

```python
data.isnull().sum()
```

```python
data.drop('Cabin', axis=1, inplace=True)
display(data)
```

```python
data.describe()
```

```python
data[['Age', 'Fare']] = data[['Age', 'Fare']].fillna(data[['Age', 'Fare']].mean())

data.describe()
```

```python
data.hist()
```

```python
def gender_cleaning(x):
    if x=="male":
        return "M"
    else:
        return "F"

data["Sex"]= data["Sex"].apply(gender_cleaning)
data.head()
```

```python
grouped = data.groupby('Survived').size()
```

```python
gender_survival = data.groupby('Sex')['Survived'].count()
gender_survival
```

```python
data[data['Survived']==1].count()
```

```python
data[data['Survived']==0].count()
```

```python
def seperate_names(x):
    first_name = x.split(',')[1].strip()
    last_name = x.split(',')[0].strip()
    return first_name, last_name

data[['First Name', 'Last Name']]= data['Name'].apply(seperate_names).apply(pd.Series)
data.drop('Name', axis = 1, inplace=True)
display(data)
```

```python
class_survival = data[data['Survived']==1].groupby('Pclass')['Survived'].count()
print(class_survival)

class_groups = data.groupby('Pclass').size()
print(class_groups)
```

```python
class_survival.plot(kind='pie',autopct='%1.1f%%', legend = True)
plt.title('Survival of people by their class basis')
plt.show()
```

```python
def age_section(x):
    if x>18 and x<=40:
        return 'A'
    elif x<18:
        return 'Y'
    else:
        return 'O'
data[['Age Division']] = data['Age'].apply(age_section).apply(pd.Series)
display(data)
```

```python
total_age = data.groupby('Age Division').size()
total_age.plot(kind='pie', legend = True, autopct='%1.1f%%', labels = [f'{x} {y}' for x, y in zip(total_age.index, total_age)])
plt.title('Total Age Distribution')
plt.show()

age_dist = data[data['Survived']==1].groupby('Age Division')['Survived'].count().plot(kind='pie', legend=True, autopct='%1.1f%%')
plt.title('Age wise distribution of Survivals')
plt.show()
```

```python
embarked_group = data.groupby('Embarked').size()
embarked_group.plot(kind='pie', autopct='%1.1f%%', labels = [f'{x} {y}' for x, y in zip(embarked_group.index, embarked_group)])
plt.title('Embarked from which ports')
plt.show()

embarked_survived = data[data['Survived']==1].groupby('Embarked').size()
embarked_survived.plot(kind='pie', autopct='%1.1f%%', labels = [f'{x} {y}' for x, y in zip(embarked_survived.index, embarked_survived)])
plt.title('Survivals from different embarked ports')
plt.show()
```

```python
ports_fare_range = data.groupby('Embarked')['Fare'].agg(lambda x: x.max() - x.min())

print(ports_fare_range)
# Plotting the range of fares for different ports
ports_fare_range.plot(kind='bar', figsize=(10, 6), color='orange')

plt.title('Range of Fare for People Embarked from Different Ports')
plt.xlabel('Embarkation Port')
plt.ylabel('Fare Range')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig('range_of_fare.png', format='png', dpi=300)
plt.show()
```

```python
mean_fares = data.groupby('Embarked')['Fare'].mean()
data['Fares'] = data.apply( lambda row: mean_fares[row['Embarked']] if row['Fare']==0 else row['Fare'] , axis=1)
display(data)
```
